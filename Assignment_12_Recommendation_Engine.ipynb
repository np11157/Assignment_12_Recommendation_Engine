## Problem statement - Build a recommender system by using cosine simillarties score.

```python
import pandas as pd
```

```python
book_data = pd.read_csv('book (1).csv')
book_data
```

```python
book_data_1 = book_data.rename({'User.ID':'User_ID',
                          'Book.Title':'Book_Title',
                          'Book.Rating':'Book_Rating'},axis=1)
book_data_1
```

```python
book_data_1.Book_Title.unique()
```

```python
book_data_1.User_ID.unique()
```

```python
book_data_1.User_ID.nunique()
```

```python
book_data_1.isna().sum()
```

```python
# Conversion of long data in other formate
user_recomm = pd.pivot_table(data = book_data_1, values='Book_Rating',index='User_ID',columns='Book_Title').fillna(0)
user_recomm
```

```python
# Replacing the index value with unique User Id
user_recomm.index=book_data_1.User_ID.unique()
user_recomm
```

```python
user_recomm.values
```

```python
from sklearn.metrics import pairwise_distances
```

```python
# Cosine Similarity for Recommendation Engine
user_to_user_cosine = 1-pairwise_distances(user_recomm.values, metric='cosine',)
```

```python
user_to_user_cosine_df = pd.DataFrame(data=user_to_user_cosine, index=book_data_1['User_ID'].unique(), columns=book_data_1['User_ID'].unique())
user_to_user_cosine_df
```

```python
#Nullify diagonal values
import numpy as np
np.fill_diagonal(user_to_user_cosine_df.to_numpy(), val = 0)
user_to_user_cosine_df
```

```python
# Most similar user
user_to_user_cosine_df.idxmax()
```
